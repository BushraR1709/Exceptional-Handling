Q1. Explain why we have to use the Exception class while creating a Custom Exception

```python
# While creating a custom exception we use Exception class because exception class is a base class and custom exception 
# inherit all the properties of Exception class to handle all type of errors. Exception classes allows us to 
# create a hierarchy of exceptions, making it easier to organize and categorize different types of errors in our code. 
```

Q2. Write a python program to print Python Exception Hierarchy.

```python
def exception_hierarchy(exception_class, indent=0):
    print('  ' * indent + exception_class.__name__)
    for subclass in exception_class.__subclasses__():
        exception_hierarchy(subclass, indent + 1)

print("Python Exception Hierarchy:")
exception_hierarchy(BaseException)

```
